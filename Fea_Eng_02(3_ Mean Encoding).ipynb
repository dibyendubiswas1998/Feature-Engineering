**To** perform **mean** **encoding**, you need to check
is your output feature **categorical** data or **neumerical** data.
If output features is categorical data then perform some **encoding** **techniques** to convert **neumerical** data.

```python
# convert to neumerical data (output feature)
def Encoding(data, output_feature, dicts):
  data[output_feature] = data[output_feature].map(dicts)

Encoding(data, 'output', {'grate':5, 'very good':4, 'good':3, 'okay':2, 'not okay':1})
data.head(5)
```

```python
print(dict(data.groupby(['fea2'])['output'].mean()))
print(dict(data.groupby(['fea3'])['output'].mean()))
print(dict(data.groupby(['state'])['output'].mean()))
```

```python
def Mean_Encoding(data, featureCol, outputCol):
  mean_nominal = data.groupby([featureCol])[outputCol].mean().to_dict()
  data[featureCol+'_mean_encoding'] = data[featureCol].map(mean_nominal)
  # data.drop(axis=1, columns=featureCol, inplace=True)
```

```python
Mean_Encoding(data, 'fea2', 'output')
Mean_Encoding(data, 'fea3', 'output')
Mean_Encoding(data, 'state', 'output')
```
