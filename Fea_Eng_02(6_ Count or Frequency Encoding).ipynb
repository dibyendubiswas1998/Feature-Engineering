```python
def get_numberOf_record(data):
  for col in data.columns:
    print(f"{col}:  {data[col].unique()} and have {len(data[col].unique())} categories")

get_numberOf_record(data)
```

```python
def frequency_Encoding(data, feature_col, output_col):
  frequency = dict(data.groupby([feature_col])[output_col].count().sort_values(ascending=False))
  data[feature_col+'_fre_encoding'] = data[feature_col].map(frequency)
  print(feature_col, ':', frequency)
  
```

```python
frequency_Encoding(data, 'fea2', 'output')
frequency_Encoding(data, 'fea3', 'output')
frequency_Encoding(data, 'state', 'output')
```
