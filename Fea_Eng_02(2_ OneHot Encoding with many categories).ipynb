```python
def get_numberOf_record(data):
  for col in data.columns:
    print(f"{col}:  {data[col].unique()} and have {len(data[col].unique())} categories")

get_numberOf_record(data)
```

Here I take top 3 categories from **fea2**, **fea3**, **state** features.

```python
def get_numberOf_records_in_features(data, col):
  records = dict(data[col].value_counts().sort_values(ascending=False))
  print(col,":",  records)


get_numberOf_records_in_features(data, 'fea2')
get_numberOf_records_in_features(data, 'fea3')
get_numberOf_records_in_features(data, 'state')
```

```python
def take_top3_records(data, col):
  return data[col].value_counts().sort_values(ascending=False).head(3).index

top3_fea2 = take_top3_records(data, 'fea2')
top3_fea3 = take_top3_records(data, 'fea3')
top3_state = take_top3_records(data, 'state')
```

```python
print(top3_fea2)
print(top3_fea3)
print(top3_state)
```

```python
# perform encoding:
def Encoding(data, top_categories, col):
  for category in top_categories:
    data[col+'_'+category] = np.where(data[col] == category, 1,0)
  # data.drop(axis=1, columns=col, inplace=True)
```

```python
Encoding(data, top3_fea2, 'fea2')
Encoding(data, top3_fea3, 'fea3')
Encoding(data, top3_state, 'state')
```
