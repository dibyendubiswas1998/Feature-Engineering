**To** perform Probabiliy ratio encoding based on **'fea2'**, **'fea3'**, **'state'**  **features**.

```python
def Prob_Ratio_Encoding(data, feature_col, output_col, p1, p2):
  positive = data.groupby([feature_col])[output_col].mean()
  negative = 1 - data.groupby([feature_col])[output_col].mean()
  
  data[feature_col+'_'+p1] = data[feature_col].map(dict(positive))
  data[feature_col+'_'+p2] = data[feature_col].map(dict(negative))
  data[feature_col+'_prob_ratio'] = data[feature_col+'_'+p1] / data[feature_col+'_'+p2]

  data.drop(axis=1, columns=[feature_col+'_'+p1, feature_col+'_'+p2], inplace=True)
```

```python
Prob_Ratio_Encoding(data, 'fea2', 'status', 'positive', 'negative')
Prob_Ratio_Encoding(data, 'fea3', 'status', 'positive', 'negative')
Prob_Ratio_Encoding(data, 'state', 'status', 'positive', 'negative')
```
