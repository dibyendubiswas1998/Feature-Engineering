```python
def get_numberOf_record(data):
  for col in data.columns:
    print(f"{col}:  {data[col].unique()} and have {len(data[col].unique())} categories")

get_numberOf_record(data)
```

```python
data['output'] = data['output'].map({'grate':5, 'very good':4, 'good':3, 'okay':2, 'not okay':1})
```

```python
def target_guided_encode(data, col, output_col):
  labels = data.groupby([col])[output_col].mean().sort_values().index
  dicts = {k:i for i,k in enumerate(labels, 0)}
  data[col+'_tg_encode'] = data[col].map(dicts)
  # data.drop(axis=1, columns=col, inplace=True)
  
```
