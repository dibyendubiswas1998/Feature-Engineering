```python
missing_vals = ['Na', 'NA', 'NOO', 'NO']
df = pd.read_csv('dataset.csv', na_values=missing_vals)
df
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(), yticklabels=False, annot=True)
plt.show()
```

## If All the columns have null or nan value then Drop it

```python
data_drop = df
data_drop.head(2)
```

```python
data_drop.dropna(how='all', inplace=True)
data_drop
```

## Fill the NAN value with zero

```python
df_fill = df
df_fill.fillna(0, inplace=True) #filling with zero is not a good approaches
df_fill
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
df['b'].plot(kind='kde', ax=ax)
df_fill.b.plot(kind='kde', ax=ax, color='red')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
plt.show()
```

## Fill The NAN values with ffill, bfill

```python
df = pd.read_csv('dataset.csv', na_values=missing_vals)
df_ffill = df
df_ffill.fillna(method='ffill', inplace=True)
df_ffill
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
df['b'].plot(kind='kde', ax=ax)
df_ffill.b.plot(kind='kde', ax=ax, color='red')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
plt.show()
```

## Replace the NAN value with mean, median and mode:

```python
df = pd.read_csv('dataset.csv', na_values=missing_vals)
df_mean = df
```

```python
df_mean['a'].fillna(df_mean.a.mean(), inplace=True)
df_mean['c'].fillna(df_mean.c.mean(), inplace=True)
df_mean['d'].fillna(df_mean.d.mean(), inplace=True)
df_mean
```

```python
df = pd.read_csv('dataset.csv', na_values=missing_vals)
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
df['b'].plot(kind='kde', ax=ax)
df_mean.b.plot(kind='kde', ax=ax, color='red')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
plt.show()
```

## Random Sample Imputation

```python
df =  pd.read_csv('dataset.csv', na_values=missing_vals)
df
```

```python
df['a'].dropna().sample()
```

```python
df['a'].dropna().sample(df['a'].isnull().sum(),random_state=0)
```

```python
df['a_random_impute'] = df['a']
random_sample = df['a'].dropna().sample(df['a'].isnull().sum(),random_state=0)
random_sample.index = df[df['a_random_impute'].isnull()].index
df.loc[df['a_random_impute'].isnull(), 'a_random_impute'] = random_sample
df
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
df['a'].plot(kind='kde', ax=ax)
df['a_random_impute'].plot(kind='kde', ax=ax, color='red')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
plt.show()
```
